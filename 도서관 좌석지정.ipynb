```python
menu = '''[안녕하세요. 좌석지정 프로그램을 시작합니다.]
------------------------------
- 열람실 1 35번 ~ 41번
- 열람실 2 78번 ~ 97번
- 열람실 3 42번 ~ 49번

1. 동 2. 호수 4. 휴대폰 뒷4자리 5. 입주민 여부 확인 6. 열람실 선택 7. 좌석 선택
위 순서대로 진행됩니다.
------------------------------'''
print(menu)
data1 = int(input("어느 동에 거주하세요?"))
data2 = int(input("몇 호에 거주하세요?"))
data3 = int(input("휴대폰 뒷 4자리를 입력해주세요"))

# 개인정보 동의 확인
consent = input("개인정보 수집 및 이용에 동의하십니까? (yes/no): ")

if consent.lower() == "yes":
  print("개인정보 수집 및 이용에 동의하셨습니다. 열람실을 사용하실 수 있습니다.")
else:
  print("개인정보 수집 및 이용에 동의하지 않으셨습니다. 열람실을 사용하실 수 없습니다.")

// 입주민 파일과 비교
import pandas as pd
# 엑셀 파일 읽어오기
df = pd.read_excel('파일명.xlsx')  # 파일명을 실제 파일명으로 변경

# 특정 열에서 데이터가 모두 존재하는지 확인
if df[(df['동'] == data1) & (df['호수'] == data2) & (df['휴대폰'] == data3)].empty == False:
  print("안녕하세요 입주민님! 원하시는 열람실을 선택해주세요")
else:
  print("입주민이 아니시네요. 사용이 불가합니다")

# 각 공간별 좌석 정보
spaces = {
    "1": list(range(1, 8)),  # 열람실1 공간: 1~8번 좌석
    "2": list(range(9, 37)), # 열람실2 공간: 9~36번 좌석
    "3": list(range(37, 45))  # 열람실3 공간: 37~44번 좌석
}
  while True:
    # 공간 선택
    selected_space = input("좌석을 지정할 공간을 선택하세요 (1, 2, 3): ").upper()

    if selected_space in spaces:
       # Check if the selected room has available seats
        if not available_seats[selected_space]:  # Check if the list is empty
            print(f"{selected_space} 열람실에는 남은 자리가 없습니다. 다른 열람실을 선택해주세요.")
            continue  # Go back to the beginning of the loop to select another room

      current_seats = available_seats[selected_space]

      while current_seats:
        display_available_seats(current_seats)

        try:
          selected_seat = int(input("좌석 번호를 선택하세요: "))

          # 좌석 지정 및 결과 확인
          if assign_seat(current_seats, selected_seat):
            break  # 좌석 지정 성공 시 현재 공간 루프 종료

        except ValueError:
          print("잘못된 입력입니다. 빈 좌석번호를 입력해주세요.")

      else:
        print(f"{selected_space} 공간의 모든 좌석이 지정되었습니다.")
        break  # 현재 공간의 모든 좌석이 지정되면 전체 루프 종료

    else:
      print("잘못된 열람실을 선택했습니다. 다시 선택해주세요.")
  print("모든 좌석이 지정되었습니다.")

// 남은 자리  List  보여주기
def display_available_seats(seats):
  """남은 좌석을 출력하는 함수"""
  print("해당 열람실 좌석:")
  for seat in seats:
    print(seat, end=" ")
  print()

def assign_seat(seats, selected_seat):
  """좌석을 지정하고 남은 좌석을 업데이트하는 함수"""
  if selected_seat in seats:
    seats.remove(selected_seat)
    print(f"{selected_seat}번 좌석이 지정되었습니다.")
  else:
    print("이미 사용중인 좌석입니다. 다른 좌석을 선택해주세요.")

# 전체 좌석
total_seats = list(range(70, 97))  # 70번부터 97번까지 좌석

# 초기 남은 좌석 (전체 좌석으로 시작)
available_seats = total_seats[:]

while available_seats:
  display_available_seats(available_seats)

  try:
    selected_seat = int(input("좌석 번호를 선택하세요: "))
    assign_seat(available_seats, selected_seat)

  except ValueError:
    print("잘못된 입력입니다. 좌석번호를 입력해주세요.")

print("모든 좌석이 지정되었습니다.")
```
